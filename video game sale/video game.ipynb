```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
num_col = data.select_dtypes(include='number').drop(['index','Year_of_Release'],axis=1)
```

```python
column_name = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']
```

```python
num_col.head()
```

```python
plt.figure(figsize=(10,8))
plotnum = 1
for i in range(len(column_name)):
    plt.subplot(5,1,plotnum)
    sns.boxplot(x =column_name[i],data=data)
    plt.tight_layout()
    plotnum += 1
plt.show()
```
